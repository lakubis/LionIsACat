```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import matplotlib

matplotlib.style.use('ggplot')
sns.set_style("whitegrid")
```

```python
df=pd.read_csv('Simulasi Less Station.csv')
df.drop(df.columns[0], axis = 1, inplace=True)
df.describe().iloc[0]
```

```python
df[df.columns[0]].count()
```

```python
cols_ = []
for i in df.columns :
    if df[i].count() == 30: 
        cols_.append(i)
df_final = df[cols_]
```

```python
c = 5
aspect = [16, 9]
fig = plt.figure(figsize=(aspect[0]*c,aspect[1]*c))
df_final.boxplot()
plt.title('Box Plot Optimization')
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
df_final[df_final.columns[:10]].boxplot()
```

```python
fig = plt.figure(figsize=(12,6))
df_final[df_final.columns[10:20]].boxplot()
```
